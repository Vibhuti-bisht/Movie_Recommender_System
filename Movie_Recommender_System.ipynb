## Helper function

#### first convert string of lists to list with the help of ast module

```python
import ast
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 
```

###  ast.literal_eval

```python
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}]')
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

#### top-three actors

```python
def convert3(obj):
    L = []
    counter=0
    for i in ast.literal_eval(obj):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L
    
```

```python
movies['cast'] = movies['cast'].apply(convert)
```

```python
movies.head(1)
```

```python
movies['cast'] = movies['cast'].apply(lambda x:x[0:3])
```

```python
movies['crew'][0]
```

```python
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew']= movies['crew'].apply(fetch_director)
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
print(stem('loved'))
```

```python
new_df['tags']= new_df['tags'].apply(stem)
```

### now we calculate the cosine distance between the  movie vectors and check the degree between the vectors. 180 degree means completely opposite movies

```python
#distance is inversely proportional to similarity
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

# Finding only top 5 similar movies

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movie_list = sorted(list(enumerate(distances)),reverse=True, key = lambda x:x[1])[1:6]
    
    for i in movie_list:
        print(new_df.iloc[i[0]].title)
        
```
